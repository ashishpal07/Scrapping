```python
import requests
import re

r = requests.get('https://www.keeplaughingforever.com/funny-riddles').text

pattern = re.compile(r'<p class="font_8" style="font-size:20px"><span style="font-size:20px"><span style="font-family:basic,sans-serif">(.*?)</span></span></p>')

matches = pattern.finditer(r)

for match in matches:
    riddle = match.group(1)
    
    riddle = re.sub('<.*?>', '', riddle, flags=re.S)
    
    print(riddle)
```
